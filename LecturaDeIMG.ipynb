```python
from pathlib import Path
import cv2
import matplotlib.pyplot as plt
import numpy as np

cwd = Path.cwd()
TRAIN_PATH = cwd / 'train_zip' /'train'
images = list(TRAIN_PATH.glob('*.jpg'))

images = [image.relative_to(cwd) for image in images]

img = cv2.imread(str(images[0]),cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```
