```python
from pathlib import Path
import cv2
import matplotlib.pyplot as plt
import numpy as np
from read_xml import XmlDictConfig
from xml.etree import cElementTree as ElementTree

cwd = Path.cwd().parents[0]
TRAIN_PATH = cwd / 'train_zip' / 'train'

images = list(TRAIN_PATH.glob('orange*.jpg'))

images = [image.relative_to(cwd) for image in images]

img = cv2.imread('../'+str(images[10]),cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
images_box = list(TRAIN_PATH.glob('orange*.xml'))
images_box = [image_box.relative_to(cwd) for image_box in images_box]
```

```python
tree = ElementTree.parse('../'+str(images_box[10]))
root = tree.getroot()
# Read Boxes dict
boxes=[]
for index ,box in enumerate(root.findall('object')):
    print(box.find('bndbox').find('xmin').text)
    boxes.append({
                 name:box.find('bndbox').find(name).text for name in ['xmin','ymin','xmax','ymax']
                 })
boxes
```

```python
pos={key:int(value) for key, value in boxes[0].items()}
pos
```

```python
from matplotlib.patches import Rectangle

# cv2.rectangle(img, start_point, end_point,(255,0,0), 2)
rect = Rectangle((pos['xmin'],pos['ymin']),
                  pos['xmax']-pos['xmin'],
                  pos['ymax']-pos['ymin'],
                 linewidth=5,edgecolor='r',facecolor='none')
plt.gca().add_patch(rect)
plt.imshow(img)
```

```python
for box in boxes:
    pos={key:int(value) for key, value in box.items()}
    rect = Rectangle((pos['xmin'],pos['ymin']),
                      pos['xmax']-pos['xmin'],
                      pos['ymax']-pos['ymin'],
                     linewidth=3,edgecolor='g',facecolor='none')
    plt.gca().add_patch(rect)
plt.imshow(img)
```
